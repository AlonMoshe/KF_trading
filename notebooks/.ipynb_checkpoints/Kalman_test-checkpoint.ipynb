***Imports***

```python
# === 1. Imports ===
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

**Getting data**

After getting daily data from IB API, the following is the processing.

```python


# === 2. Load the data ===
file_path = r"C:\Users\User\OneDrive\Projects\TradingSystem\TWSAPI\Data\TSLA_data.csv"
df = pd.read_csv(file_path)

# === 3. Convert 'Date' column to timezone-aware datetime ===
# The format in your file looks like: '20251103 09:30:00 US/Eastern'
# We'll parse manually and localize to US/Eastern timezone
df['Datetime'] = pd.to_datetime(
    df['Date'].str.replace(' US/Eastern', ''),  # remove the timezone string
    format="%Y%m%d %H:%M:%S"
).dt.tz_localize('US/Eastern')

# Set as index for time series operations
df = df.set_index('Datetime').sort_index()

# === 4. Compute derived quantities ===
# Mid-price (average of high and low)
df['Mid'] = (df['Open'] + df['Close']) / 2

# Proxy for spread (High - Low)
df['Spread'] = df['High'] - df['Low']

Delta = 5.0  # seconds
lambda_ = 0.94  # decay factor for EWMA variance

# EWMA variance of mid-price changes (used as volatility proxy)

returns = df['Mid'].diff()
df['var_dMid'] = (returns**2).ewm(alpha=(1 - lambda_),adjust=False).mean()
df['var_acc'] = df['var_dMid'] / (Delta**4)
df['Spread_smooth'] = df['Spread'].ewm(alpha=0.2, adjust=False).mean()


# === 5. Clean up ===-*0
df = df[['Open', 'High', 'Low', 'Close', 'Volume', 'Mid',
         'Spread', 'Spread_smooth', 'var_dMid', 'var_acc']]


# === 6. Basic inspection ===
print(df.head())
print("\nRows:", len(df))
print("Time range:", df.index[0], "→", df.index[-1])

# === 7. Optional: Save preprocessed data for reuse ===
save_path = r"C:\Users\User\OneDrive\Projects\TradingSystem\TWSAPI\Data\TSLA_preprocessed.csv"
df.to_csv(save_path)
print(f"\nPreprocessed data saved to: {save_path}")
```

**Kalman filter**

```python


# We assume df already exists from your preprocessing step and has:
# index = timezone-aware DatetimeIndex
# columns: ['Open','High','Low','Close','Volume','Mid','Spread','Volatility']

# Safety check
required_cols = {'Mid','Spread','var_dMid'}
missing = required_cols - set(df.columns)
if missing:
    raise ValueError(f"DataFrame 'df' is missing required columns: {missing}")
else:
    print("All is good")
```

```python
# === 2) Model setup (local quadratic: level, slope, curvature) ===


A = np.array([
    [1.0,       Delta,        0.5*(Delta**2)],
    [0.0,       1.0,          Delta],
    [0.0,       0.0,          1.0]
], dtype=float)

H = np.array([[1.0, 0.0, 0.0]], dtype=float)  # observe 'level' only (Mid ≈ level)


```

```python
A,H
```

```python
# === 3) Process/measurement noise scaffolding ===
# Constant-acceleration process noise (driven by acceleration variance q_acc)
# Discrete-time Q for white acceleration noise is standard:

def Q_ca(q_acc: float, Delta: float) -> np.ndarray:
    d, d2, d3, d4 = Delta, Delta**2, Delta**3, Delta**4
    return q_acc * np.array([
        [d4/4.0, d3/2.0, d2/2.0],
        [d3/2.0, d2,     d     ],
        [d2/2.0, d,      1.0   ]
    ], dtype=float)

# --- RETUNE NOISE SCALES FOR VISIBLE SMOOTHING ---
half_spread_med = (df['Spread'] / 2.0).median()
r_meas = max(1e-8, float(half_spread_med**2))

mid = df['Mid'].values
dmid = np.diff(mid, prepend=mid[0])
typical_dmid = np.median(np.abs(dmid[1:]))

# ↓↓↓ MUCH SMALLER Q (process), BIGGER R (measurement)
q_scale = 1e-4      # was ~1e-2…1; make it MUCH smaller
r_scale = 50.0      # make measurement noise bigger

q_acc = max(1e-12, float((typical_dmid / Delta)**2) * q_scale)
R_fixed = np.array([[r_meas * r_scale]], dtype=float)
Q_fixed = Q_ca(q_acc, Delta)

print("q_acc:", q_acc, " | R_fixed:", R_fixed[0,0])
```

```python
Q_fixed
```

```python
# === 4) Initialization ===
y = df['Mid'].to_numpy(dtype=float)

# x represents the state vector
# initial state: level=start price, slope=0, curvature=0
x = np.zeros((3, 1), dtype=float)
x[0, 0] = y[0]

# initial covariance: high uncertainty on slope/curvature; moderate on level
P = np.diag([half_spread_med**2, (typical_dmid**2), 1.0]) * 100.0

I3 = np.eye(3, dtype=float)

# containers
n = len(y)
level_est  = np.zeros(n)
slope_est  = np.zeros(n)
curv_est   = np.zeros(n)
level_var  = np.zeros(n)
slope_var  = np.zeros(n)
curv_var   = np.zeros(n)
gain_track = np.zeros(n)
```

```python
x,P,I3, len(level_est)
```

```python
# === 5) Kalman filter (baseline: fixed Q,R) ===
for t in range(n):
    # Predict
    x = A @ x
    P = A @ P @ A.T + Q_fixed

    # Update
    z = np.array([[y[t]]])                      # observation
    S = H @ P @ H.T + R_fixed                   # innovation covariance
    K = P @ H.T @ np.linalg.inv(S)              # Kalman gain
    innov = z - (H @ x)                         # innovation

    x = x + K @ innov
    P = (I3 - K @ H) @ P

    # Save
    level_est[t] = x[0, 0]
    slope_est[t] = x[1, 0]
    curv_est[t]  = x[2, 0]
    level_var[t] = P[0, 0]
    slope_var[t] = P[1, 1]
    curv_var[t]  = P[2, 2]
    gain_track[t] = K[0,0]

# Attach to DataFrame
df['KF_level_fixed'] = level_est
df['KF_slope_fixed'] = slope_est
df['KF_curv_fixed']  = curv_est
df['KF_level_var_fixed'] = level_var
df['KF_slope_var_fixed'] = slope_var
df['KF_curv_var_fixed']  = curv_var

print("Baseline Kalman (fixed Q,R) complete.")
```

```python
# === 6) Adaptive Q_t, R_t (optional, recommended after you see baseline) ===
# R_t scales with spread; Q_t scales with EWMA volatility
# These multipliers are tunable knobs:
cR = 10   # measurement noise scale vs. spread^2
cQ = 1e-4   # process noise scale vs. volatility^2

# --- Reasonable floors to avoid tiny R_t / huge responsiveness ---
R_floor = (df['Spread'].median()/2.0)**2 * 5.0   # baseline measurement noise
Q_floor = q_acc * 0.2                             # re-use your fixed q_acc as scale ref

x = np.zeros((3, 1), dtype=float)
gain_adapt = np.zeros(n)
x[0, 0] = y[0]
P = np.diag([half_spread_med**2, (typical_dmid**2), 1.0]) * 100.0

level_est_a  = np.zeros(n)
slope_est_a  = np.zeros(n)
curv_est_a   = np.zeros(n)
level_var_a  = np.zeros(n)
slope_var_a  = np.zeros(n)
curv_var_a   = np.zeros(n)

# === Vectorized arrays for speed ===
var_acc_arr = df['var_acc'].to_numpy()
spread_arr  = df['Spread_smooth'].to_numpy()
y_arr       = y

for t in range(n):

    # Predict
    x = A @ x

    # Adaptive Q_t
    q_t = var_acc_arr[t]
    Q_t = Q_ca(max(Q_floor, cQ * q_t), Delta)

    # Adaptive R_t (smoothed spread)
    spr = spread_arr[t]
    R_t = np.array([[max(R_floor, cR * (spr**2))]], dtype=float)

    # Covariance prediction
    P = A @ P @ A.T + Q_t

    # Measurement update
    z = np.array([[y_arr[t]]])
    S = H @ P @ H.T + R_t
    K = P @ H.T @ np.linalg.inv(S)
    innov = z - (H @ x)
    x = x + K @ innov
    P = (I3 - K @ H) @ P

    # Save
    level_est_a[t] = x[0, 0]
    slope_est_a[t] = x[1, 0]
    curv_est_a[t]  = x[2, 0]
    level_var_a[t] = P[0, 0]
    slope_var_a[t] = P[1, 1]
    curv_var_a[t]  = P[2, 2]
    gain_adapt[t] = K[0,0]

df['KF_level_adapt'] = level_est_a
df['KF_slope_adapt'] = slope_est_a
df['KF_curv_adapt']  = curv_est_a
df['KF_level_var_adapt'] = level_var_a
df['KF_slope_var_adapt'] = slope_var_a
df['KF_curv_var_adapt']  = curv_var_a

print("Adaptive Kalman (Q_t,R_t) complete.")
print("Mean adaptive Kalman gain:", gain_adapt[gain_adapt>0].mean())
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from ipywidgets import Output
from IPython.display import display

# pick a short sample window
window = 4000
df_sample = df.iloc[:window]

fig = make_subplots(
    rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.03,
    subplot_titles=("Price", "Slope", "Curvature")
)

# --- Panel 1: Price (Mid + KF fixed + KF adapt) ---
fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['Mid'],
    mode='lines+markers', name='Mid',
    line=dict(color='gray', width=1), marker=dict(size=3)
), 1, 1)

fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['KF_level_adapt'],
    mode='lines+markers', name='KF Level (Adaptive)',
    line=dict(color='royalblue', width=1.5), marker=dict(size=3)
), 1, 1)

fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['KF_level_fixed'],
    mode='lines+markers', name='KF Level (Fixed Q,R)',
    line=dict(color='darkorange', width=2), marker=dict(size=3, color='darkorange')
), 1, 1)

# --- Panel 2: Slope (fixed + adapt) ---
fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['KF_slope_adapt'],
    mode='lines+markers', name='Slope (Adaptive)',
    line=dict(color='darkorange', width=1.5), marker=dict(size=3)
), 2, 1)

fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['KF_slope_fixed'],
    mode='lines+markers', name='Slope (Fixed Q,R)',
    line=dict(width=1.5, color='red'), marker=dict(size=3)
), 2, 1)

# --- Panel 3: Curvature (fixed + adapt) ---
fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['KF_curv_adapt'],
    mode='lines+markers', name='Curvature (Adaptive)',
    line=dict(color='forestgreen', width=1.5), marker=dict(size=3)
), 3, 1)

fig.add_trace(go.Scatter(
    x=df_sample.index, y=df_sample['KF_curv_fixed'],
    mode='lines+markers', name='Curvature (Fixed Q,R)',
    line=dict(color='firebrick', width=1.5), marker=dict(size=3)
), 3, 1)

# --- Layout ---
fig.update_layout(
    height=900, width=1100, template='plotly_white',
    hovermode='x',  # standard x-hover; we'll control the vertical line via a shape
    legend=dict(x=0.01, y=0.99),
    title='Kalman Filter: Price, Slope and Curvature'
)

for i, label in enumerate(['Price', 'Slope', 'Curvature'], start=1):
    fig.update_yaxes(title_text=label, row=i, col=1)
fig.update_xaxes(title_text='Time', row=3, col=1)

# --- Add a single vertical line spanning all panels (initially hidden) ---
x0 = df_sample.index[0]
fig.add_shape(
    type='line', xref='x', yref='paper',
    x0=x0, x1=x0, y0=0, y1=1,
    line=dict(color='red', width=1),
    visible=False  # hidden until hover
)

# --- Interactive update: show line on hover, hide on unhover ---
out = Output()
display(out)

def move_vline(trace, points, state):
    if points.xs:
        x_new = points.xs[0]
        # move and show the line
        fig.update_shapes(dict(x0=x_new, x1=x_new, visible=True))
        with out:
            out.clear_output(wait=True)
            display(fig)

def hide_vline(trace, points, state):
    # hide the line when cursor leaves a trace
    fig.update_shapes(dict(visible=False))
    with out:
        out.clear_output(wait=True)
        display(fig)

# attach callbacks to any one trace (or multiple if you like)
fig.data[0].on_hover(move_vline)
fig.data[0].on_unhover(hide_vline)

fig.show()
```
