```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)          # Function 1
y2 = np.exp(0.3 * x)    # Function 2

# Create the main figure and axis
fig, ax1 = plt.subplots()

# Plot first function (left y-axis)
ax1.plot(x, y1, color='tab:blue', label='sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

# Plot second function (right y-axis)
ax2.plot(x, y2, color='tab:red', label='exp(0.3x)')
ax2.set_ylabel('exp(0.3x)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Optional: add a title and grid
plt.title('Two functions with separate y-axes')
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(0.3 * x)

# Create figure
fig = go.Figure()

# Left y-axis trace
fig.add_trace(go.Scatter(
    x=x, y=y1,
    mode='lines',
    name='sin(x)',
    line=dict(color='blue')
))

# Right y-axis trace
fig.add_trace(go.Scatter(
    x=x, y=y2,
    mode='lines',
    name='exp(0.3x)',
    line=dict(color='red'),
    yaxis='y2'  # <- Assign to second y-axis
))

# Define the two y-axes
fig.update_layout(
    title='Two functions with separate y-axes (interactive)',
    xaxis=dict(title='x'),
    yaxis=dict(title='sin(x)', color='blue'),
    yaxis2=dict(
        title='exp(0.3x)',
        color='red',
        overlaying='y',   # share same x-axis
        side='right'      # position on right
    ),
    legend=dict(x=0.05, y=0.95),
    hovermode='x unified'
)

fig.show()
```
